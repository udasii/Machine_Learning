# Problem Statement:

In the landscape of the oil and gas industry, the ability to accurately forecast the productivity of wells is challenging for companies seeking to optimize their production rights investment decisions. The primary challenge revolves around the current forecasting methods. Presently, the industry relies heavily on the manual creation of 'type curves' derived from historical data to predict future production trends of similar wells. However, the effectiveness of this approach is limited, failing to leverage the full potential of available data.

The overarching goal is to maximize return on investment by identifying wells with the highest potential for profitability. Achieving this objective necessitates enhancing the accuracy of production forecasts, thereby mitigating risks and capitalizing on lucrative opportunities. 

- The importance lies in identifying the underlying patterns within the dataset to identify key parameters that drive productivity. We have considered Well Type, Perforation Interval Level, and Production Stage to classify well into distinct categories. Based on these factors, dynamic forecasting models enable stakeholders to make well-informed decisions, leading to optimized resource allocation and maximized returns.

**Approach:** 
The approach entails a comprehensive analysis of the available data through exploratory data analysis (EDA) to unveil critical variables influencing production. This involves cleaning data, identifying significant features, and formulating predictive models.

**Steps:**
- Conduct thorough data cleaning procedures to ensure the integrity of the datasets.
- Undertake exploratory data analysis to gain insights into the distribution and characteristics of key variables.
- Select relevant variables essential for forecasting models based on statistical metrics and domain expertise.
- Implement advanced clustering techniques, such as k-means clustering, to categorize variables and identify influential factors.
- Develop forecasting models, utilizing techniques like SARIMAX along with grid search and exogenous factors such as Water Production to predict future production trends.
- Validate and refine the models based on accuracy metrics such as AIC, ensuring robustness and reliability.

# Assumptions made


In the pursuit of meaningful insights, certain assumptions were necessitated:

1. Interpretation of Unexplained Variables:
Due to a lack of comprehensive explanations for some variables, assumptions were made regarding their meanings. For instance, all instances of "EUR" were assumed to represent "Expected Ultimate Recover," akin to the concept of recoverable reserves.

2. Variable Relevance and Analyst's Intuition:
Given the constrained time frame and the analyst's limited expertise in the oil & gas sector, variables impactful on well production were selected intuitively. This includes geographical details and well characteristics.

3. Limitation to Specific Data Sets:
Acknowledging the time constraints, the analysis assumed the aditional data sets provided besides the 2 used, were irrelevant for the study. 

5. Focus on Producing Wells:
For the purpose of this analysis, producing wells were exclusively considered, excluding other statuses assumed to be less favorable for investment due to their non-producing nature.

6. Uniformity Across Wells: 
It is assumed that the categorization of wells based on lateral length is representative of similar operational conditions and geological characteristics. This assumes that all wells within a category are comparable and that lateral length is a significant independent variable influencing the KPIs.

7. Median as a Representative Statistic: 
The use of median values to represent the central tendency of the data assumes that the median is an adequate measure of central tendency for the distribution of each KPI, even though this may not account for skewed distributions or outliers.

8. Stable Technological and Operational Conditions: 
The analysis presumes that the technological and operational conditions under which the wells operate remain constant over time. It assumes that advancements in technology or changes in operational techniques have not significantly impacted the performance metrics across different production stages or lateral lengths.

9. Consistency in Decline Rates: 
It is assumed that historical decline rates, as observed in the provided data, will continue into the future, allowing the type curves to be predictive. This assumes that the geological and operational conditions remain consistent over time for similar wells.

10. Representativeness of the Sample: 
The data used to generate the type curves is assumed to be representative of the wider population of oil and gas wells. This means the wells in the dataset are considered a good proxy for similar wells not included in the analysis. It is assumed that past production trends are reliable indicators of future performance. This implies that historical cumulative production data used to create type curves will continue to accurately forecast future production rates.

# Problem Solving

```python
# Selecting features
features = df_filtered_md[['PerfInterval_FT', 'ENVWellType']]

# Encoding categorical data
features = pd.get_dummies(features, columns=['ENVWellType'])

# Standardizing the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
inertia = []
for k in range(1, 11):  # Test 1 to 10 clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features_scaled)
    inertia.append(kmeans.inertia_)

# Plotting the elbow curve
plt.figure(figsize=(8, 4))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Determining K')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

- Elbow Method Analysis: Utilizes the KMeans clustering algorithm to identify the optimal number of clusters (k) by plotting the inertia (within-cluster sum of squares) against different k values (1 to 10). The 'elbow point' in the graph, where the inertia begins **(at 4)** to decrease more slowly, indicates the most appropriate number of clusters for our dataset.

```python
# k=4 is chosen based on the elbow method
k = 4  
kmeans = KMeans(n_clusters=k, random_state=42)
df_filtered_md['Cluster'] = kmeans.fit_predict(features_scaled)
```

### Cluster Class 2: Perforation Interval Length

PerfInterval_FT is relevant for oil and gas wells because it provides essential information about the characteristics of the reservoir and plays a significant role in determining production efficiency.

```python
# Select relevant columns
perfIntervals = df_filtered_md[['PerfInterval_FT']]

# Specify the number of clusters (4 for Short, Medium, Long, Extra Long)
num_clusters = 4

# Apply KMeans clustering
kmeans = KMeans(n_clusters=num_clusters, n_init=10, random_state=42)
df_filtered_md['PerfInterval_Cluster'] = kmeans.fit_predict(perfIntervals)

# Map cluster labels
perint_mapping = {
    0: 'Short',
    1: 'Medium',
    2: 'Long',
    3: 'Extra Long'
}

df_filtered_md['Cluster_PerfInterval'] = df_filtered_md['PerfInterval_Cluster'].map(perint_mapping)

# Scatter plot for clusters
plt.figure(figsize=(10, 6))
for cluster in range(num_clusters):
    cluster_data = df_filtered_md[df_filtered_md['PerfInterval_Cluster'] == cluster]
    plt.scatter(cluster_data['PerfInterval_FT'], cluster_data['CumProd_BOEPer1000FT'], label=f'Cluster {cluster}')

# Add labels and title
plt.xlabel('PerfInterval_FT')
plt.ylabel('CumProd_BOEPer1000FT')
plt.title('Clusters of PerfInterval_FT')
plt.legend()
plt.show()
```

```python
# Define the range of values for p, d, q, P, D, Q, and m
p_values = range(0, 3)  # Autoregressive order
d_values = [0]          # Differencing order
q_values = range(0, 3)  # Moving average order
P_values = range(0, 2)  # Seasonal autoregressive order
D_values = range(0, 1)  # Seasonal differencing order
Q_values = range(0, 2)  # Seasonal moving average order
m_values = [12]         # Seasonal period

# Create all possible combinations of SARIMA parameters
param_combinations = list(itertools.product(p_values, 
                                            d_values, 
                                            q_values, 
                                            P_values, 
                                            D_values, 
                                            Q_values, 
                                            m_values))

# Initialize AIC with a large value
best_aic_oil = float("inf")  
best_params_oil = None

# Perform grid search
for params_oil in param_combinations:
    order = params_oil[:3]
    seasonal_order = params_oil[3:]
    
    try:
        model_oil = sm.tsa.SARIMAX(df_train_oil, 
                        order=order, 
                        seasonal_order=seasonal_order)
        result_oil = model_oil.fit(disp=False)
        aic_oil = result_oil.aic
        
        # Ensure the convergence of the model
        if not math.isinf(result_oil.zvalues.mean()):
            print(order, seasonal_order, aic_oil)
        
            if aic_oil < best_aic_oil:
                best_aic_oil = aic_oil
                best_params_oil = params_oil
                
        else:
            print(order, seasonal_order, 'not converged')

    except:
        continue

# Print the best parameters and AIC
print("Best Parameters Oil:", best_params_oil)
print("Best AIC Oil:", best_aic_oil)
```
